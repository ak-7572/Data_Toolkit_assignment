```python
# Q3. Explain the following terms
# The difference in np.array, np.asarray and np.asanyarrayX
# The difference between Deep copy and shallow copyX


# Differences between np.array, np.asarray, and np.asanyarray

# np.array:
# Creates a new array from any object (lists, tuples, etc.).
# Always returns a new array object. This means it will copy the data from the input object.
# Use when you want to ensure that you're working with a new array, regardless of the input type.
# Example:
import numpy as np
a = [1, 2, 3]
arr1 = np.array(a)


# np.asarray:
# Converts the input to an array but does not make a copy if the input is already an array.
# If the input is already a NumPy array, it returns the same object. Otherwise, it copies the data from the input object.
# Use when you want to minimize copying and avoid unnecessary data duplication.
# Example:
b = np.array([1, 2, 3])
arr2 = np.asarray(b)
arr3 = np.asarray([1, 2, 3])



# np.asanyarray:
# Similar to np.asarray, but it allows for subclasses of ndarray (like matrix).
# If the input is a subclass of ndarray (like a matrix), it returns that subclass instead of converting it to a base ndarray.
# Use when you want to retain subclasses (e.g., matrices) in the output.
# Example:
c = np.matrix([[1, 2], [3, 4]])
arr4 = np.asanyarray(c)
arr5 = np.asanyarray([1, 2, 3])



# Difference between Deep Copy and Shallow Copy

# Shallow Copy:
# A shallow copy creates a new object, but inserts references into it to the objects found in the original. If the original object contains nested objects (like lists within lists), the shallow copy does not create copies of the nested objects; it only copies their references.
# Changes made to the nested objects in the original will reflect in the shallow copy and vice versa.
# Example:
import copy

original = [[1, 2, 3], [4, 5, 6]]
shallow_copied = copy.copy(original)

shallow_copied[0][0] = 'X'
print(original)



# Deep Copy:

# A deep copy creates a new object and recursively copies all objects found in the original, meaning that nested objects are also copied.
# Changes made to the nested objects in the original will not affect the deep copy and vice versa.
# Example:
import copy

original = [[1, 2, 3], [4, 5, 6]]
deep_copied = copy.deepcopy(original)

deep_copied[0][0] = 'X'
print(original)
```

```python
# Q6. Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:
# a) Find the indices of the maximum values along each depth level (third axis).
# b) Perform element-wise multiplication of between both arrayX


import numpy as np

array_3d = np.random.randint(1, 11, size=(3, 3, 3))

print("3D Array:")
print(array_3d)

max_indices = np.argmax(array_3d, axis=2)
print("\nIndices of Maximum Values Along Each Depth Level:")
print(max_indices)

elementwise_multiplication = array_3d * array_3d
print("\nElement-wise Multiplication of the Array with Itself:")
print(elementwise_multiplication)
```

```python
# Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
# b) Label the x-axis as 'Date'.
# c) Set the title of the plot as 'Temperature and Humidity Over Time'.


import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'Temperature': [30, 32, 31, 29, 28, 30, 34, 33, 32, 31],
    'Humidity': [60, 65, 55, 70, 68, 65, 72, 75, 80, 78]
}

df = pd.DataFrame(data)

fig, ax1 = plt.subplots()

ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Temperature and Humidity Over Time')

plt.show()
```

```python
# Q15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.
# b) Overlay a line plot representing the normal distribution's probability density function (PDF).
# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
# d) Set the title of the plot as 'Histogram with PDF Overlay'.


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.randn(1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

mean = np.mean(data)
std_dev = np.std(data)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, color='blue', label='Normal PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

plt.title('Histogram with PDF Overlay')

plt.legend()

plt.grid()
plt.show()
```

```python
# Q17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

data = pd.DataFrame({'X': x, 'Y': y})

def get_quadrant(row):
    if row['X'] >= 0 and row['Y'] >= 0:
        return 'Quadrant 1'
    elif row['X'] < 0 and row['Y'] >= 0:
        return 'Quadrant 2'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

data['Quadrant'] = np.where((data['X'] >= 5) & (data['Y'] >= 5), 'Quadrant 1',
                     np.where((data['X'] < 5) & (data['Y'] >= 5), 'Quadrant 2',
                     np.where((data['X'] < 5) & (data['Y'] < 5), 'Quadrant 3', 'Quadrant 4')))

plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='Set1', s=100)

scatter_plot.set_xlabel('X-axis')
scatter_plot.set_ylabel('Y-axis')

plt.title('Quadrant-wise Scatter Plot')

plt.legend(title='Quadrants')

plt.grid()
plt.show()
```

```python
# Q18.  With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function

from math import pi
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = np.linspace(0, 2 * pi, 100)
y = np.sin(x)

p = figure(title='Sine Wave Function', x_axis_label='X (radians)', y_axis_label='Y (sin(x))', width=800, height=400)

p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')

output_notebook()
show(p)
```

```python
# Q19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'8


import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
values = np.random.randint(1, 100, size=len(categories))

data = pd.DataFrame({'Category': categories, 'Value': values})

color = ['#%02x%02x%02x' % (int(value * 2.55), 0, 255 - int(value * 2.55)) for value in values]

source = ColumnDataSource(data=data)

p = figure(x_range=data['Category'], title='Random Categorical Bar Chart',
           x_axis_label='Category', y_axis_label='Value',
           toolbar_location=None, tools='')

p.vbar(x='Category', top='Value', source=source, width=0.9, color=color)

hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

p.xgrid.grid_line_color = None
p.y_range.start = 0

output_notebook()
show(p)
```

```python
# Q20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = np.random.rand(100)

trace = go.Scatter(
    x=x,
    y=y,
    mode='lines',
    name='Random Data',
    line=dict(color='blue', width=2)
)

layout = go.Layout(
    title='Simple Line Plot',
    xaxis=dict(title='X-axis'),
    yaxis=dict(title='Y-axis')
)

fig = go.Figure(data=[trace], layout=layout)

pyo.iplot(fig)
```

```python
# Q21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo

np.random.seed(42)
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(labels))

pie_chart = go.Figure(data=[go.Pie(
    labels=labels,
    values=values,
    textinfo='label+percent',
    hoverinfo='label+percent',
    marker=dict(line=dict(color='#000000', width=1)))])

pie_chart.update_layout(title='Interactive Pie Chart')

pyo.iplot(pie_chart)
```
